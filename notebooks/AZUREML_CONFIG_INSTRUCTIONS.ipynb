# AzureML Workspace Setup Instructions

To use AzureML in this notebook, you need a `config.json` file with your workspace credentials.

**How to get config.json:**
1. Go to [Azure ML Studio](https://ml.azure.com).
2. Click your workspace name in the top right.
3. Select 'Download config.json'.
4. Place the file in your project root directory (`energy-ai-azure`).

Once `config.json` is in place, rerun the AzureML setup cell in your main notebook.

If you want to use a different location, provide the path to `Workspace.from_config(path="/path/to/config.json")`.
